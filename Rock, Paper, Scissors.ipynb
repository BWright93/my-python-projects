```python
import random

# What if we wish to play multiple games in a row instead of one at a time
# We can create a while loop and add it to our code above
# The match ends when either the player or computer reach a certain score

print(f'''
=================================
Welcome to Rock, Paper, Scissors!
=================================
''') # Introductory message

player_score = 0 # Keep track of how many wins both the player and computer have
computer_score = 0
play_again = ['Y','y']
end_game = ['N','n']

player_name = input('Please enter your name: ') # Ask for name which'll be displayed throughout the game

while True:
    
    # Create an input function for the player
    player_turn = input('\nSelect Rock, Paper or Scissors: ')

    # List of different choices
    computer_choices = ['Rock','Paper','Scissors']
    # Set the computers turn to randomly choose from the list given
    computer_turn = random.choice(computer_choices)

    # We can also 'print/display' the choices made by both the player and computer as follows
    print(f"\n{player_name} chose: {player_turn.capitalize()}, \nComputer chose: {computer_turn}\n")

    # To determine the winner, we can create an if/elif/else block to compare both choices
    if player_turn.capitalize() == computer_turn:
        print('Draw!\n') # Both player and computer choices match therefore its a draw
    elif player_turn.capitalize() == 'Rock':
        if computer_turn == 'Scissors':
            print('Rock beats Scissors! Player Wins!\n')
            player_score += 1 # Add one point to the players score
        else:
            print('Paper beats Rock! Computer Wins!\n')
            computer_score += 1 # Add one point to the computers score
    elif player_turn.capitalize() == 'Paper':
        if computer_turn == 'Rock':
            print('Paper beats Rock! Player Wins!\n')
            player_score += 1
        else:
            print('Scissors beats Paper! Computer Wins!\n')
            computer_score += 1
    elif player_turn.capitalize() == 'Scissors':
        if computer_turn == 'Paper':
            print('Scissors beats Paper! Player Wins!\n')
            player_score += 1
        else:
            print('Rock beats Scissors! Computer Wins!\n')
            computer_score += 1
    else: # If the player types an invalid option
        print('Invalid option! Please try again!\n')

    print('The current scores are:\n')
    print('' + player_name + ' has '+str(player_score)+' wins.')
    print('Computer has '+str(computer_score)+' wins.\n')

# Create 'if' loops for when the player or computer reach a certain score

    if player_score == 3: 
        print(f'''
===================================
Game Over! {player_name} wins the match!
===================================
        ''')
        another_game = input('Would you like to play again? Y/N: ')
        if another_game in play_again:
            player_score = 0
            computer_score = 0
            continue
        else:
            print('Thanks for playing!')
            break
    
    if computer_score == 3:
        print(f'''
===================================
Game Over! Computer wins the match!
===================================
        ''')
        another_game = input('Would you like to play again? Y/N: ')
        if another_game in play_again:
            player_score = 0
            computer_score = 0
            continue
        else:
            print(f'''
===================
Thanks for playing!
===================
            ''')
            break # Sets the 'while' loop to False and Ends the game"
```
