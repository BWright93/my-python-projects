Creating a function that generates a random password of a specific length given by the user.

```python
def password(num):

    import string
    import random

    char_list = ''

    while(True):
        # Create input variable to decide which characters we want for our password
        # Added 'or 4' as default value in case of empty user input
        pass_choice = int(input('Letter, number or special? (1-3) Press any other key to exit.') or 4)

        # Add letters to character list
        if pass_choice == 1:
            char_list += string.ascii_letters
        # Add numbers to character list
        elif pass_choice == 2:
            char_list += string.digits
        # Add special characters to character list
        elif pass_choice == 3:
            char_list += string.punctuation
        else:
            break

    # We cannot create a random password from an empty character list
    # Raise an custom error for this scenario
    if len(char_list) == 0:
        raise Exception('Character list empty! Password cannot be generated!')

    password = []

    # Create a 'for' loop which chooses random characters from our list
    # Then adds them to our password
    for i in range(num):

        random_char = random.choice(char_list)

        password.append(random_char)

    print('\nHere is your password')
    return ''.join(password)

i = password(int(input('Choose password length: ')))

if len(i) == 0:
    raise Exception('No password length given. Please try again!')
else:
    print(i)
```
