```python
import random
from random import shuffle

game_list = ['Heads','Tails']

def coin_flip(n): # Flip/toss a coin 'n' times and counts how many heads and tails we get
    
    assert n > 0 
    
    heads_count = 0 # Placeholders
    tails_count = 0
    
    for x in range(n):
        flip = random.choice(game_list)
        if flip == 'Heads':
            heads_count += 1
            print('Heads')
        else:
            tails_count += 1
            print('Tails')
    print('\nNumber of heads: ' + str(heads_count))
    print('Number of tails: ' + str(tails_count))
    if heads_count > tails_count:
        print('Heads wins!')
    elif tails_count > heads_count:
        print('Tails wins!')
    else:
        print('Its a draw!')

def main():  # Wrapper function
    # Create an input function for the user to choose how many coin flips they want
    # Using the 'coin flip' function we just created

    coin_flip(int(input('How many flips? ')))

if __name__ == '__main__':
    main()
```
