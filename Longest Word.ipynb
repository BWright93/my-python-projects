Checking which is the longest word in a string excluding punctuation. <br>
Numbers are included when checking each word. <br>
If there are two or more longest words of equal length then the first one in the string is selected.

```python
import string

def LongestWord(sen):
  assert len(sen) > 0
  for char in string.punctuation:
    sen = sen.replace(char, ' ')
  words = sen.split()
  wor = ''
  for word in words:
    if len(word) > len(wor):
      wor = word
  return wor

another = ('Y','y')

while True:
    l = input('Insert sentence here: ')
    print(f'The longest word is: "{LongestWord(l)}"')
    print(f'The longest word has {len(LongestWord(l))} letters.')
    another_word = input('Wanna try again? ')
    if another_word in another:
        continue
    else:
        print('Goodbye!')
        break
```
