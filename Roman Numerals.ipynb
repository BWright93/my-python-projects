Create a function that converts Roman Numerals into their numerical value.

```python
# First create a function which sets each Roman Numeral to it's numerical value

def r_value(rnum):

    valid_chars = {'I','V','X','L','C','D','M'}
    for char in rnum.upper():
        if char not in valid_chars:
            return 'Invalid Input'
            break
    
    if rnum.upper() == 'I':
        return 1
    if rnum.upper() == 'V':
        return 5
    if rnum.upper() == 'X':
        return 10
    if rnum.upper() == 'L':
        return 50
    if rnum.upper() == 'C':
        return 100
    if rnum.upper() == 'D':
        return 500
    if rnum.upper() == 'M':
        return 1000
    return 'No Roman Numeral given.'

# Now create a function which sums up all Roman Numerals in a string correctly
# Must take into account for smaller numbers preceding larger numbers, ie. IV = 4, XC = 90
def r_sum(str):

    # Empty placeholder for our value result
    result = 0
    i = 0

    # Check if we've reached the end of the string containing our Roman Numeral
    while (i < len(str)):

        if str[i].upper() not in {'I','V','X','L','C','D','M'}:
            return 'Invalid input! Please try again!'
            break

        # Set first string value to correct numerical value using our r_value function above
        str_value1 = r_value(str[i])

        if ((i+1) < len(str)):

            # Set second string value to correct numerical value
            str_value2 = r_value(str[i+1])

            # Add first string value to our 'result' variable if it's larger or equal than second string value
            if (str_value1 >= str_value2):

                result = result + str_value1
                i = i + 1

            # Otherwise add second string value first then subtract first string value
            else:

                result = result + str_value2 - str_value1
                i = i + 2

        # If we only have one Roman Numeral letter
        else:
            result = result + str_value1
            i = i + 1

    # Returns the numerical value equivalent to our Roman Numerical given
    return result

try_again = ('Y','y')

while True:
    
    total_sum = r_sum(input('Enter Roman Numeral: '))
    print('The numerical value of is', total_sum)
    next_string = input('Wanna try again? Press Y/y. ')
    if next_string in try_again:
        continue
    else:
        print('Goodbye!')
        break
```
