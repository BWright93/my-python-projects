Create a working calculator using Tkinter.

```python
import tkinter as tk
from tkinter import messagebox
```

```python
expr = '' # Global expression value

# Function for whenever a button is pressed
def press(key):
    global expr
    expr += str(key)
    display.set(expr)

# Function to change between positive or negative number
def flip():
    global expr
    if '-' in expr:
        f_num = expr.replace('-','')
    else:
        f_num = '-' + expr
    display.set(f_num)
    expr = f_num

# Function to get the result when '=' button is pressed
def equal():
    global expr
    try:
        result = str(eval(expr)) # Use 'eval' to evaluate the string and return the result
        display.set(result)
        expr = ''
    except:
        display.set('Error!') # Displays error message if string does not evaluate correctly
        expr = ''

# Function to clear the display when 'C' button is pressed
def clear():
    global expr
    expr = ''
    display.set('')

# Function that'll close the calculator
def Close():
    root.destroy()

# Create a tkinter window, set the title and size/geometry
root = tk.Tk()
root.title('Basic Tkinter Calculator')
root.configure(bg='white')
#root.geometry('500x500')

label = tk.Label(root, text='Welcome!',font=('arial',20))
label.grid(row=0,column=1,columnspan=4)

# Create a display for the calculator sums
display = tk.StringVar()
first_entry = tk.Entry(root,textvariable=display,font=('arial',16))
first_entry.grid(row=1,column=1,columnspan=4)

# Create buttons for each number 0-9 and basic math symbols

endbutton = tk.Button(root, text='C',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=clear)
endbutton.grid(row=2,column=1,pady=0)

endbutton = tk.Button(root, text='(',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15), command=lambda: press('('))
endbutton.grid(row=2,column=2,pady=0)

endbutton = tk.Button(root, text=')',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15), command=lambda: press(')'))
endbutton.grid(row=2,column=3,pady=0)

endbutton = tk.Button(root, text='sqrt',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15), command=lambda: press('**0.5'))
endbutton.grid(row=2,column=4,pady=0)

endbutton = tk.Button(root, text='7', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(7))
endbutton.grid(row=3,column=1,pady=0)

endbutton = tk.Button(root, text='8', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(8))
endbutton.grid(row=3,column=2,pady=0)

endbutton = tk.Button(root, text='9', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(9))
endbutton.grid(row=3,column=3,pady=0)

endbutton = tk.Button(root, text='/', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press('/'))
endbutton.grid(row=3,column=4,pady=0)

endbutton = tk.Button(root, text='4',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(4))
endbutton.grid(row=4,column=1,pady=0)

endbutton = tk.Button(root, text='5',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(5))
endbutton.grid(row=4,column=2,pady=0)

endbutton = tk.Button(root, text='6',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(6))
endbutton.grid(row=4,column=3,pady=0)

endbutton = tk.Button(root, text='*', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press('*'))
endbutton.grid(row=4,column=4,pady=0)

endbutton = tk.Button(root, text='1',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(1))
endbutton.grid(row=5,column=1,pady=0)

endbutton = tk.Button(root, text='2',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(2))
endbutton.grid(row=5,column=2,pady=0)

endbutton = tk.Button(root, text='3',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(3))
endbutton.grid(row=5,column=3,pady=0)

endbutton = tk.Button(root, text='-', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press('-'))
endbutton.grid(row=5,column=4,pady=0)

endbutton = tk.Button(root, text='0',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press(0))
endbutton.grid(row=6,column=1,pady=0)

endbutton = tk.Button(root, text='.',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press('.'))
endbutton.grid(row=6,column=2,pady=0)

endbutton = tk.Button(root, text='^',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15), command=lambda: press('**'))
endbutton.grid(row=6,column=3,pady=0)

endbutton = tk.Button(root, text='+', bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=lambda: press('+'))
endbutton.grid(row=6,column=4,pady=0)

endbutton = tk.Button(root, text='=',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15),command=equal)
endbutton.grid(row=7,column=2,pady=0)

endbutton = tk.Button(root, text='+/-',bg='blue',fg='white',width=5,height=2,font=('Helvetica',15), command=flip)
endbutton.grid(row=7,column=3,pady=0)

# Close the calculator window
endbutton = tk.Button(root, text='Close',bg='red',fg='white', command=Close)
endbutton.grid(row=7,column=4,pady=0)


root.mainloop()
```
