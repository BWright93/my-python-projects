Check if a given string is a palindrome (Reads the same backwards as forwards ie. racecar). <br>
Whitespace and special characters ie. Uppercase/Lowercase & Punctuations are ignored

```python
def isPalindrome(s):

    # Change entire string to lowercase (or uppercase if you prefer)
    s = s.lower()

    # Empty placeholder for new string 
    new_s = ''

    # Create a 'for' loop to check every character in the string
    for i in s:

        # Check if character 'i' is in the alphabet, if so then adds/concatenates 'i' into the new string 'new_s'
        if i.isalpha():
            new_s += i

    # Checks if the new string is identical to its reversed counterpart, then returns either True or False
    return new_s == new_s[::-1]

try_again = ('Y','y')

# Create a 'while' loop to check multiple strings if necessary
while True:
    string = input('Insert string here: ')
    result = isPalindrome(string)
    print(result)
    next_string = input('Wanna try again? Press Y/y. ')
    if next_string in try_again:
        continue
    else:
        print('Ending program!')
        break
```
