```python
import math

def Factorial(num):
    result = 1
    if num == 0:
        return 1
    else:
        for i in range(1,num+1):
            result *= i
        return result

try_again = ('Y','y')

while True:
    n = int(input('Please input number here: '))
    if n < 0:
        print('Cannot input negative number!\n')
    else:
        print(f"The factorial for this number is  {Factorial(n)}\n")
    another_num = input('Would you like to try another number? If yes type Y')
    if another_num in try_again:
        continue
    else:
        print('Goodbye!')
        break
```
