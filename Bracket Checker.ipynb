Check that a given string has brackets/parenthesis that match up.

```python
def BracketChecker(strParam):

    # Empty placeholder variable we'll use to count the brackets
    bracket_count = 0

    # Create a 'for' loop which'll check through the given string
    for i in strParam:
        # Check for the left parenthesis then check for its right corresponding one afterward
        if i == '(':
            bracket_count += 1
        elif i == ')':
            bracket_count -= 1
            # End check if there's a right parenthesis first, before a left one
            if bracket_count == -1:
                return "Error! Left parenthesis missing!\n"

    if bracket_count == 0:
        return "No parenthesis found or they match up correctly!\n"
    else:
        return "Incorrect parenthesis layout!\n"

# Create a 'while' loop that'll allow us to check multiple strings

start_over = ('Y', 'y')

while True:
    bracketstr = input('Insert string here: ')
    if bracketstr == "":
        print('No input given!\n')
    else:
        check = BracketChecker(bracketstr)
        print(check)
    new_string = input('Another string to check? Press Y/y to continue. ')
    if new_string in start_over:
        continue
    else:
        print('Thank you & goodbye!')
        break
```
